```python
search_btn.click()
```

```python
# fetching the product url

urls = []
start = 0
end = 3
for page in range(start,end):
    tit = driver.find_elements(By.XPATH,'//a[@class="a-link-normal s-underline-text s-underline-link-text s-link-style a-text-normal"]')
    for i in tit:
        urls.append(i.get_attribute('href'))
        
    next_bu = driver.find_elements(By.XPATH,"/html/body/div[1]/div[2]/div[1]/div[1]/div/span[3]/div[2]/div[65]/div/div/span/a[3]")
 
```

```python
len(urls)
```

```python
urls
```

```python
# fetching the name of the product

names = []

for i in urls:
    driver.get(i)
    time.sleep(5)
    try:
        pro = driver.find_element(By.XPATH,'/html/body/div[1]/div[2]/div[7]/div[6]/div[4]/div[1]/div/h1/span')
        names.append(pro.text)
        
    except NoSuchElementException :
        names.append('Not Present')
 
```

```python
names
```

```python
# fetching the price of the product

price = []

for i in urls:
    driver.get(i)
    time.sleep(3)
    
    
    try:
        pros = driver.find_element(By.XPATH,'/html/body/div[1]/div[2]/div[6]/div[4]/div[1]/div[4]/div/div/div/div/form/div/div/div/div/div[2]/div[1]/div/span/span[2]/span[2]')
        price.append(pros.text)
        
    except NoSuchElementException :
        price.append('Not Present')
```

```python
price
```

```python
# fetching the return/exchange of the product

returns = []

for i in urls:
    driver.get(i)
    time.sleep(3)
    
    
    try:
        pr = driver.find_element(By.XPATH,'/html/body/div[1]/div[2]/div[7]/div[6]/div[1]/div[4]/div/div/div/div/form/div/div/div/div/div[3]/div/div[21]/div/div/span/span/a')
        returns.append(pr.text)
        
    except NoSuchElementException :
        returns.append('Not Present')
```

```python
returns
```

```python
# fetching the availability of the product

aval = []

for i in urls:
    driver.get(i)
    time.sleep(3)
    
    
    try:
        prose = driver.find_element(By.XPATH,'/html/body/div[1]/div[2]/div[6]/div[4]/div[1]/div[4]/div/div/div/div/form/div/div/div/div/div[3]/div/div[4]/div/div[1]/span')
        aval.append(prose.text)
        
    except NoSuchElementException :
        aval.append('Not Present')
```

```python
aval
```

```python
# fetching the expected delivery of the product

deli = []

for i in urls:
    driver.get(i)
    time.sleep(3)
    
    
    try:
        p = driver.find_element(By.XPATH,'/html/body/div[1]/div[2]/div[9]/div[6]/div[1]/div[4]/div/div/div/div/form/div/div/div/div/div[3]/div/div[2]/div[8]/div[1]/div/div/div[2]/span')
        deli.append(p.text)
        
    except NoSuchElementException :
        deli.append('Not Present')
```

```python
deli
```

```python
data = pd.DataFrame({'NAMES':names,'PRICE':price[0:192],'RETURN':returns[0:192],'AVAILABILITY':aval[0:192],'DELIVERY':deli[0:192]})
data 
```

```python
# fetching the hostel names

nam = driver.find_elements(By.XPATH,'//h2[@class="title title-6"]')
len(nam)
```

```python
names = []

for i in nam:
    names.append(i.text)
    
names
```

```python
hostel_url = []



while(True):

    urls = driver.find_elements(By.XPATH,'//h2[@class="title title-6"]//a')

    for url in urls:

        hostel_url.append(url.get_attribute("href"))

    time.sleep(2)    

        

    try:

        next_button = driver.find_elements(By.XPATH,'//div[@class="pagination-item pagination-next"]')

        next_button.click()

    except:

        break
```

```python
# fetching the property descr

prop = []

for i in hostel_url:
    driver.get(i)
    time.sleep(4)
    try:
        pro = driver.find_element(By.XPATH,'//html//body//div[3]//div//div//div[1]//section//div[6]//div//div[2]//div[2]//div//div[2]')
        prop.append(pro.text)
        
    except NoSuchElementException :
        prop.append('Not Present')
```
