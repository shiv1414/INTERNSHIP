# Q1) Scrape the details of most viewed videos on YouTube from Wikipedia.

```python
driver = webdriver.Chrome('chromedriver.exe')
```

```python
url = "https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos"
driver.get(url)
```

```python
# EXTRACTING THE Name OF the videos

nam = driver.find_elements(By.XPATH,"//table[@class='wikitable sortable jquery-tablesorter']//tbody//tr//td[2]")
len(nam)
```

```python
name = []

for i in nam[0:30]:
    name.append(i.text)
    
name
    
    
```

```python
# EXTRACTING THE VIDEOS RANk

ran = driver.find_elements(By.XPATH,"//table[@class='wikitable sortable jquery-tablesorter']//tbody//tr//td[1]")
len(ran)
```

```python
rank = []

for i in ran[0:30]:
    rank.append(i.text)
    
rank
    
    
```

```python
# EXTRACTING THE ARTIST

art = driver.find_elements(By.XPATH,"//table[@class='wikitable sortable jquery-tablesorter']//tbody//tr//td[3]")
len(art)
```

```python
artist = []

for i in art[0:30]:
    artist.append(i.text)
    
artist
```

```python
# EXTRACTING THE upload date

dat = driver.find_elements(By.XPATH,"//table[@class='wikitable sortable jquery-tablesorter']//tbody//tr//td[5]")
len(dat)
```

```python
date = []

for i in dat[0:30]:
    date.append(i.text)
    
date
```

```python
# EXTRACTING THE views

vie = driver.find_elements(By.XPATH,"//table[@class='wikitable sortable jquery-tablesorter']//tbody//tr//td[4]")
len(vie)
```

```python
view = []

for i in vie[0:30]:
    view.append(i.text)
    
view
```

```python
df = pd.DataFrame({'RANK':rank,'NAME':name,'ARTIST':artist,'UPLOAD DATE':date,'VIEWS(IN BILLION)':view})
df
```
